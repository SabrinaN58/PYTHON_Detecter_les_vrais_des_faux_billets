```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
# Chargement du fichier 
df = pd.read_csv("billets_production.csv")
df.head()
```

```python
# Création d'une fonction pour détecter les vrais et faux billets

def prediction_billet(df):
    #Sélection des variables significatives
    X = df.drop(["diagonal","height_left", "id"], axis=1)
    #Standardisation des données
    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    #Chargement du model de régression logistique
    rl_model = joblib.load("selected_model")
    #Probabilités et prédiction
    df["proba_0"] = rl_model.predict_proba(X)[:,0]
    df["proba_1"] = rl_model.predict_proba(X)[:,1]
    prediction = rl_model.predict(X)
    #Ajout des prédictions dans le dataframe
    df["prediction"] = prediction

    #Dataframe contenant les prédictions
    return df
```

```python
#Visualisation des prédictions

prediction_billet(df)
```

Les billets A_1, A_2 et A_3 sont prédits comme faux\
Les billets A_4 et A_5 sont prédits comme vrais
